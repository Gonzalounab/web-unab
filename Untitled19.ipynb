```python
import yfinance as yf
import pandas as pd
import requests
from bs4 import BeautifulSoup

# Descargar datos bursátiles de Tesla (TSLA) utilizando yfinance
tesla_data = yf.download("TSLA", start="2020-01-01", end="2023-01-01")

# Restablecer el índice y mostrar las primeras 5 filas
tesla_data.reset_index(inplace=True)
print("Datos bursátiles de Tesla (primeras 5 filas):")
print(tesla_data.head())

# URL para obtener los ingresos de Tesla desde MacroTrends
url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"

# Configurar los encabezados para la solicitud HTTP
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3"
}

# Hacer la solicitud a la página web
try:
    response = requests.get(url, headers=headers)
    response.raise_for_status()  # Verificar si la solicitud fue exitosa
except requests.exceptions.RequestException as e:
    print("Error al hacer la solicitud:", e)
    exit()

# Analizar el contenido HTML de la página
soup = BeautifulSoup(response.content, "html.parser")

# Verificar si hay tablas disponibles en la página
tables = soup.find_all("table")
if len(tables) < 2:
    print("No se encontraron suficientes tablas en la página.")
    exit()

# Extraer la tabla de ingresos
revenue_table = tables[1]

# Convertir la tabla HTML a un DataFrame de pandas
try:
    tesla_revenue = pd.read_html(str(revenue_table))[0]
    tesla_revenue.dropna(inplace=True)  # Eliminar filas vacías
except ValueError as e:
    print("Error al leer la tabla HTML:", e)
    exit()

# Mostrar las últimas 5 filas del DataFrame de ingresos
print("\nIngresos de Tesla (últimas 5 filas):")
print(tesla_revenue.tail())
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Descargar datos bursátiles de Tesla (TSLA) usando yfinance
tesla_data = yf.download("TSLA", start="2020-01-01", end="2023-01-01")

# Restablecer el índice
tesla_data.reset_index(inplace=True)

# Función para graficar
def make_graph(data, title):
    plt.figure(figsize=(12, 6))
    plt.plot(data['Date'], data['Close'], label='Cierre')
    plt.xlabel('Fecha')
    plt.ylabel('Precio de Cierre (USD)')
    plt.title(title)
    plt.legend()
    plt.grid()
    plt.show()

# Graficar los datos de Tesla con título
make_graph(tesla_data, "Gráfico del Precio de Cierre de Tesla (2020-2023)")
```
